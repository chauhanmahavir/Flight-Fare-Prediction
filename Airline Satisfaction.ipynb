#### Feature transformation

```python
import scipy.stats as stat
import pylab
```

```python
def plot_data(df,feat):
    plt.figure(figsize=(12,6))
    plt.subplot(1,2,1)
    df[feat].hist()
    plt.subplot(1,2,2)
    stat.probplot(df[feat] , dist='norm' , plot=pylab)
    plt.show()
```

```python
plot_data(df,'Flight Distance')
```

#### RandomidsedSearchCV

```python
from sklearn.ensemble import RandomForestClassifier
from scipy.stats import randint
```

```python
RFC = RandomForestClassifier(n_jobs=-1)
```

```python
RFC_params = {
                'max_depth' : [3,5,7,10,13],
                'n_estimators' : [100,200,300,400,500],
                'criterion' : ['gini','entropy'],
                'bootstrap' : [True,False],
                'min_samples_leaf' : randint(1,4),
                'random_state' : [0,1,5,42]
                }
```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
def RMCV_Forest(Estim , params , no_iter , x , y):
    RMSearch = RandomizedSearchCV(
                                    estimator = Estim,
                                    param_distributions = params,
                                    n_jobs=-1,
                                    n_iter=no_iter,
                                    cv=9
                                )
    RMSearch.fit(x,y)
    return RMSearch.best_params_,RMSearch.best_score_
```

```python
RF_params , RF_score = RMCV_Forest(RFC , RFC_params , 10 , X_train , Y_train)
```

```python
print(f'Random Forest Best Parameters :\n\n {RF_params}\n\n')
print(f'Random Forest Best Score :\n\n {RF_score}')
```

```python
Model_best = RandomForestClassifier(bootstrap= False, criterion= 'gini', max_depth= 13, min_samples_leaf= 1, n_estimators= 200, random_state= 42)
```

```python
Model_best.fit(X_train , Y_train)
```

```python
def score(test,pred):
    print("Precision Score : ", precision_score(test,pred))
    print("Recaal Score : ", recall_score(test,pred))
    print("Accuracy : ", accuracy_score(test,pred))
    print("\n\nClassification Report : \n\n", classification_report(test,pred))
    print("\n\nConfusion Matrix : \n", confusion_matrix(test,pred))
    print("\n\nF1 Score : ",f1_score(test,pred))

Y_pred = Model_best.predict(X_test)

score(Y_test,Y_pred)
```
